# Exploration: Gaussian Process Regression

#### Code based on https://scikit-learn.org/stable/auto_examples/gaussian_process/plot_gpr_noisy_targets.html#sphx-glr-auto-examples-gaussian-process-plot-gpr-noisy-targets-py

```python
from matplotlib import pyplot as plt
import numpy
import seaborn
from sklearn.gaussian_process import GaussianProcessRegressor
```

```python
plt.rcParams["figure.figsize"] = (20, 10)

seaborn.set(font_scale=2.5)
seaborn.set_style("whitegrid")
```

```python
num_line = numpy.atleast_2d(numpy.linspace(0, 10, 1000)).T
```

```python
def func(x):
    return x * numpy.sin(x)


def fit_and_plot(x, X, func_):
    X = numpy.atleast_2d(X).T
    y = func_(X)
    
    gpr = GaussianProcessRegressor()
    gpr.fit(X, y)

    y_pred, sigma = gpr.predict(x, return_std=True)
    
    plt.scatter(X, y, s=500, c='red', label='observations')
    plt.plot(x, func_(x), color='black', linestyle='-', linewidth=6, label='$f(x) = xsin(x)$')
    plt.plot(x, y_pred, color='blue', linestyle='--', linewidth=6, label='prediction')
    plt.xlabel('$x$')
    plt.ylabel('$f(x)$')
    plt.legend(loc='upper left')
    plt.show()
```

```python
fit_and_plot(x=num_line, X=[3., 9.], func_=func)
```

```python
fit_and_plot(x=num_line, X=[3., 5., 9.], func_=func)
```

```python
fit_and_plot(x=num_line, X=[1., 2., 5., 6., 8., 9.], func_=func)
```
